```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

import cufflinks as cf
cf.go_offline()

import pandas as pd
from plotly import subplots
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
kghm = pd.read_csv('kgh_d.csv', index_col=0, parse_dates=True)
miedz = pd.read_csv('ca_c_f_d.csv', index_col=0, parse_dates=True)
```

```python
kghm.head()
```

```python
miedz.head()
```

```python
fig = subplots.make_subplots(rows=3, cols=1,
                             shared_xaxes=True,
                             subplot_titles=['KHGM', 'Mied'],specs=[[{'type':'scatter'}], [{'type':'scatter'}], [{'type':'table'}]])
fig.add_trace(go.Scatter(y=kghm['Zamkniecie']),1,1)
fig.add_trace(go.Scatter(y=miedz['Zamkniecie']),2,1)
fig.add_trace(go.Table(header=dict(values=['Data','KGHM', 'Mied']), cells=dict(values=[kghm.index.strftime('%Y-%m-%d'),kghm['Zamkniecie'], miedz['Zamkniecie']])),3,1)
fig.update_layout(height=800, showlegend=False)
iplot(fig)
```
